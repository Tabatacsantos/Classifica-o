## Balanceamento dos dados
A base de dados está desequilibrada, é necessário fazer o balanceamento para não atrapalhar o desempenho do modelo de machine learning

Método SMOTE 

SMOTE (técnica de sobreamostragem minoritária sintética) é um dos métodos de sobreamostragem mais comumente usados ​​para resolver o problema de desequilíbrio.
Seu objetivo é equilibrar a distribuição de classes, aumentando aleatoriamente os exemplos de classes minoritárias, replicando-os.
SMOTE sintetiza novas instâncias minoritárias entre instâncias minoritárias existentes. Ele gera os registros de treinamento virtual por interpolação linear para a classe minoritária. Esses registros de treinamento sintético são gerados selecionando aleatoriamente um ou mais dos k-vizinhos mais próximos para cada exemplo na classe minoritária. Após o processo de sobreamostragem, os dados são reconstruídos e vários modelos de classificação podem ser aplicados para os dados processados.

FONTE: https://acervolima.com/ml-manipulacao-de-dados-desequilibrados-com-smote-e-algoritmo-de-quase-perda-em-python/

```python
X_df= df.iloc[:, 0:30].values
X_df
```

```python
y_df = df.iloc[:, 30].values
y_df
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X_over, y_over = smote.fit_resample(X_df, y_df)
```

```python
X_over.shape
```

```python
np.unique(y_df, return_counts=True)
```

```python
np.unique(y_over, return_counts=True)
```

## Modelo Random Forest

```python
from sklearn.model_selection import train_test_split
X_df_treinamento_over, X_df_teste_over, y_df_treinamento_over, y_df_teste_over = train_test_split(X_over, y_over, test_size=0.15, random_state=0)
X_df_treinamento_over.shape, X_df_teste_over.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_df_fraude = RandomForestClassifier(criterion = 'entropy', min_samples_leaf =  1, min_samples_split = 5, n_estimators = 100)
random_forest_df_fraude.fit(X_df_treinamento_over, y_df_treinamento_over)
```

## Avaliação do modelo

```python
from sklearn.metrics import accuracy_score, classification_report
previsoes = random_forest_df_fraude.predict(X_df_teste_over)
accuracy_score(y_df_teste_over, previsoes)
```

```python
print(classification_report(y_df_teste_over, previsoes))
```
